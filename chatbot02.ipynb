```python
import google.generativeai as genai

model = genai.GenerativeModel('gemini-pro-latest')

print("Gemini model initialized successfully with 'gemini-pro-latest'.")
```

```python
chat = model.start_chat(history=[
    {
        "role": "user",
        "parts": ["From now on, consider yourself an AI assistant specialized in agriculture. How can I help you today? Please choose your role or type what you need: 1.I am a FARMER (e.g., Add a new harvest, Check my payment) 2.I am a DISTRIBUTOR (e.g., Update truck location, Verify batch quality) 3.I am a RETAILER (e.g., Show a product story, Check expiry dates) 4. I am an ADMIN (e.g., Audit temperature logs, Review system changes), providing practical advice and relevant information."],
    },
    {
        "role": "model",
        "parts": ["Understood! Hello! Welcome to the Agri-Chain Transparency System.I'm ready to assist with all your farming inquiries. What's on your mind today?"],
    },
])

print("Chatbot initialized. Type 'quit' or 'exit' to end the conversation.")

while True:
    user_input = input("You: ")
    if user_input.lower() in ['quit', 'exit']:
        print("Ending chat. Goodbye!")
        break

    response = chat.send_message(user_input)
    print(f"Gemini: {response.text}")
```
